### Importing Required Libraries

_We recommend you import all required libraries in one place (here):_


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore", category=UserWarning) 
%matplotlib inline
```

Load the dataset into a pandas dataframe


```python
filepath="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod2.csv"
df = pd.read_csv(filepath)
```

```python
# show the first 5 rows using dataframe.head() method
print("The first 5 rows of the dataframe") 
df.head(5)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.drop('Unnamed: 0.1', axis=1, inplace=True)
```

```python
df.head()
```

# Task 1 : Single Linear Regression

You have learnt that "CPU_frequency" is the parameter with the lowest p-value among the different features of the dataset. Create a single feature Linear Regression model that fits the pair of "CPU_frequency" and "Price" to find the model for prediction.


```python
# Write your code below and press Shift+Enter to execute 
lm = LinearRegression()
x = df[["CPU_frequency"]]
y = df[['Price']]
lm.fit(x, y)

yhat = lm.predict(x)
```

Generate the Distribution plot for the predicted values and that of the actual values. How well did the model perform?


```python
# Write your code below and press Shift+Enter to execute 
# Plot the actual values distribution
sns.kdeplot(df['Price'], color="r", label="Actual Value")

# Create a distribution plot for predicted values
sns.kdeplot(yhat, color="b", label="Fitted Values")

plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price')
plt.ylabel('Proportion of laptops')
plt.legend()
plt.show()
```

Evaluate the Mean Squared Error and R^2 score values for the model. 


```python
# Write your code below and press Shift+Enter to execute 
mse_slr = mean_squared_error(y, yhat)
r2_slr = lm.score(x, y)

print('The R-square for Linear Regression is: ', r2_slr)
print('The mean square error of price and predicted value is: ', mse_slr)
```

# Task 2 - Multiple Linear Regression
The parameters which have a low enough p-value so as to indicate strong relationship with the 'Price' value are 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU' and 'Category'. Use all these variables to create a Multiple Linear Regression system. 


```python
# Write your code below and press Shift+Enter to execute 
z = df[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']]

lm1 = LinearRegression()
lm1.fit(z, y)
yhat1 = lm1.predict(z)
```

Plot the Distribution graph of the predicted values as well as the Actual values


```python
# Write your code below and press Shift+Enter to execute
# Plot the actual values distribution
sns.kdeplot(df['Price'], color='r', label='Actual Values')

# Plot the predicted values distribution
sns.kdeplot(yhat1, color='b', label='Fitted Values')
plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price')
plt.ylabel('Proportion of laptops')
plt.legend()
plt.show()
```

Find the R^2 score and the MSE value for this fit. Is this better or worst than the performance of Single Linear Regression?


```python
# Write your code below and press Shift+Enter to execute
mse_mlr = mean_squared_error(y, yhat1)
r2_mlr = lm1.score(z, y)
print('The R-square for Linear Regression is: ', r2_mlr)
print('The mean square error of price and predicted value is: ', mse_mlr)
```

# Task 3 - Polynomial Regression
Use the variable "CPU_frequency" to create Polynomial features. Try this for 3 different values of polynomial degrees. Remember that polynomial fits are done using `numpy.polyfit`. 


```python
x = x.to_numpy().flatten()
y = y.to_numpy().flatten()
```

```python
#  Write your code below and press Shift+Enter to execute
f1 = np.polyfit(x,y,1)
p1 = np.poly1d(f1)
```

```python
f2 = np.polyfit(x,y,3)
p2 = np.poly1d(f2)
```

```python
f3 = np.polyfit(x,y,5)
p3 = np.poly1d(f3)
```

Plot the regression output against the actual data points to note how the data fits in each case. To plot the polynomial response over the actual data points, you have the function shown below.


```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
```

Call this function for the 3 models created and get the required graphs.


```python
#  Write your code below and press Shift+Enter to execute
# Call for function of degree 2
PlotPolly(p1, x, y, 'CPU_Frequency')
```

```python
#  Write your code below and press Shift+Enter to execute
# Call for function of degree 3
PlotPolly(p2, x, y, 'CPU_Frequency')
```

```python
#  Write your code below and press Shift+Enter to execute
# Call for function of degree 5
PlotPolly(p3, x, y, 'CPU_Frequency')
```

Also, calculate the R^2 and MSE values for these fits. For polynomial functions, the function sklearn.metrics.r2_score will be used to calculate R^2 values.


```python
#  Write your code below and press Shift+Enter to execute
r2_1 = r2_score(y, p1(x))
print('The R-square value for 1st degree polynomial is: ', r2_1)
print('The MSE value for 1st degree polynomial is: ', mean_squared_error(y,p1(x)))

r2_3 = r2_score(y, p3(x))
print('The R-square value for 3rd degree polynomial is: ', r2_3)
print('The MSE value for 3rd degree polynomial is: ', mean_squared_error(y,p2(x)))

r2_5 = r2_score(y, p3(x))
print('The R-square value for 5th degree polynomial is: ', r2_5)
print('The MSE value for 5th degree polynomial is: ', mean_squared_error(y,p3(x)))
```

# Task 4 - Pipeline
Create a pipeline that performs parameter scaling, Polynomial Feature generation and Linear regression. Use the set of multiple features as before to create this pipeline. 


```python
#  Write your code below and press Shift+Enter to execute
Input = [('scale', StandardScaler()), ('polynomial', PolynomialFeatures()), ('model', LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(z, y)
ypipe = pipe.predict(z)
```

Evaluate the MSE and R^2 values for the this predicted output.


```python
#  Write your code below and press Shift+Enter to execute
print('MSE for multi-variable polynomial pipeline is: ', mean_squared_error(y, ypipe))
print('R^2 for multi-variable polynomial pipeline is: ', r2_score(y, ypipe))
```

You should now have seen that the values of R^2 increase as we go from Single Linear Regression to Multiple Linear Regression. Further, if we go for multiple linear regression extended with polynomial features, we get an even better R^2 value.
